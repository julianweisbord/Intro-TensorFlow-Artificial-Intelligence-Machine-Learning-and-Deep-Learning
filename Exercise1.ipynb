```
# GRADED FUNCTION: house_model
def house_model(y_new):
    xs = np.array([1.0, 3.0, 9.0, 2.0, 4.0, 11.0, 5.0, 16.0, 8.0, 7.0, 6.0])  # Your Code Here#
    ys = np.array([100000.0, 200000.0, 500000.0, 150000.0, 250000.0, 600000.0, 300000.0, 850000.0, 450000.0, 400000.0, 350000.0]) # Your Code Here#
    ys = np.array([i/100000 for i in ys])
    print(ys)
    layer1 = keras.layers.Dense(units=1, input_shape=[1], activation='linear')
    model = keras.Sequential([layer1])
    model.compile(optimizer='sgd', loss='mse')
    model.fit(xs, ys, epochs=1000)
    return model.predict(y_new)[0]
```

```
prediction = house_model([7.0])
print(prediction)
```
